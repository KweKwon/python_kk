# 변수 생성규칙
* 변수는 반드시 문자로 시작해야한다.
* 변수에는 -(하이픈)을 사용할 수 없다. _(언더스코어는 가능)

```python
def dfbyid(ids):
    dfs = []
    for idx, id_name in enumerate(ids):
        globals()[f"d_{id_name}"] = over20df[over20df['ID'] == id_name] # 변수를 자동으로 생성해서 id별 dataframe만듬
        dfs.append("d_" + id_name) # id가 숫자로 시작하면 변수생성이 안되므로 앞에 'd_'를 붙임
        print(f"{idx + 1}/{len(ids)} 처리중", end='\r') # 현재 진행 사항 표시
    return dfs
```

```python
def wcBYShop(dfs):
    import string
    from emoji import core
    from wordcloud import WordCloud

    font_path = '‪C:\Windows\Fonts\malgunbd.ttf'
    # 리뷰내용을 컬럼의 리뷰를 1개의 변수에 저장하고, 텍스트 클리닝
    for val in dfs:
        test = []
        globals()[val]['리뷰내용'].apply(lambda x: test.append(x))
        test = " ".join(test)
        test.translate(str.maketrans("."," ", string.punctuation)).replace('ㅎㅎ',' ').replace('ㅋ',' ').replace('ㅋㅋ',' ').replace('ㅠㅠ',' ').replace('ㅜㅜ',' ').replace('   ',' ').replace('  ',' ')
        core.replace_emoji(test).replace('ㅎ',' ').replace('|','').replace('ㅋㅋㅋ','').replace('ㅋㅋㅋㅋ','')

        # 워드 클라우드 만들기
        wc = WordCloud(
            max_words= 200,
            font_path= font_path,
            background_color = 'white',
            stopwords = [],
            ).generate(test)
        plt.figure(figsize=(10,10))
        plt.imshow(wc)
        plt.axis('off')
        plt.savefig(f"./data/wc/{val}.png") # 저장할때 사용
        plt.close()
    print('가게별 워드클라우드 생성 완료')
```
