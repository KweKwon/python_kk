````python
tour_info = []
for p_num in data2:
    year = p_num.get('yyyymm','unkown')
    sido = p_num.get('sido','unkown')
    p_name = p_num.get('resNm','unkown')
    net_num = p_num.get('NatNum',0)
    for_num = p_num.get('ForNum','unkown')
    tour_info.append([year,sido,net_num,for_num])
    # tour_info.append([year,sido,net_num,for_num])
    # tour_info.append(temp)
````
