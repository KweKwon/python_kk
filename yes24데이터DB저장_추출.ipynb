```python
def fromyes24db(dbid, dbpw, db_name, table_name, book_id):
    from sqlalchemy import create_engine
    import pymysql
    import yes24dbinfo
    pymysql.install_as_MySQLdb()
    engine = create_engine(f"mysql+mysqldb://{dbid}:{dbpw}@192.168.0.70/{db_name}")
    conn = engine.connect() # 접속버튼누름
    yes24 = pd.read_sql(f'SELECT * FROM {table_name} WHERE book_id = {book_id}', con=conn)
    return yes24
```
