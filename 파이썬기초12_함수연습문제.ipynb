# 1. 두 개의 숫자를 입력받아 합/차/곱/나눗셈을 출력하는 calculation 함수를 작성하라.



```
calculation(3, 4)
```





```
3 + 4 = 7
3 - 4 = -1
3 * 4 = 12
3 / 4 = 0.75
```



````python
def calculation(a,b):
  print(f'{a} + {b}=', a + b)
  print(f'{a} - {b}=', a - b)
  print(f'{a} * {b}=', a * b)
  print(f'{a} / {b}=', a / b)
calculation(3,4)
````

# 2. 입력된 문자열을 역순으로 출력하는 print_reverse 함수를 정의하라.



```
print_reverse("Hello, World!")
```



```
!dlroW ,olleH
```

````python
def print_reverse(ch):
  print(ch[::-1])
print_reverse("Hello, World!")
````

# 6. 숫자로 구성된 하나의 리스트를 입력받아, 짝수들을 추출하여 리스트로 반환하는 pickup_even 함수를 구현하라.



```
pickup_even([3, 4, 5, 6, 7, 8])
```





```
[4, 6, 8]
```



````python
def pickup_even(x):
  result = ()
  for i in x:
    if i % 2 == 0:
      result.append(i)
    return result
````

````python
pickup_even([3, 4, 5, 6, 7, 8])
````
