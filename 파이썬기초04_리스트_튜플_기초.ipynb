# 리스트 정렬
.sort() : 오름차순 - 작은 숫자 -> 큰 숫자

.sort(reverse=True) : 내림차순 - 큰 숫자 -> 작은 숫자
* 실행하자마자 원본에 재할당(덮으쓰기 됨)

```python
a.sort()
```

```python
print(a)
```
