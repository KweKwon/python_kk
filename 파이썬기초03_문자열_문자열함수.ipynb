```python
h7 = "파이썬에서 줄바꿈움 \\n 으로 합니다. \n\n\n 줄 바꿈 끝"
print(h7)
```

# 문자열 함수
* 문자열을 조작하는 함수
* .split(): 특정 기호를 기준으로 문자열을 나누고 리스트로 저장
* .join(): 앞쪽에 지정한 형식을 포함해서 문자열 합쳐줌(주로 리스트의 자료를 합칠떄 사용)
* .strip(): 문자열 양쪽 끝의 공백, 특수문자를 제거
  * 문자열 내부는 제거 못함
* .replace(): 문자열에서 원하는 값을 찾아서 다른 값으로 치환해 줌.

* .strip(): 문자열 양쪽 끝의 공백, 특수문자를 제거
  * 문자열 내부는 처리하지 못함

```python
h12 = "      공백이 있는 문자열        "
h12
```

* lstrip():왼쪽 공백을 없애는 함

```python
h12.lstrip()
```

* rstrip():오른쪽 공백을 없애는 함

```python
h12.rstrip()
```

* strip():양쪽 공백을 없애줌

```python
h12.strip()
```

```python
h13 = '... , % 공백과 \n특수문자가 같이 \t있는 :문자열 !! ... $'
```

```python
h13
```

```python
h13.lstrip("., %")
```

```python
h13.rstrip("!.$")
```

```python
h13.strip(" .,%!!$")
```

* .replace("찾을문자","바꿀문자")

```python
h13.replace(".","").replace(",","").replace("!","").replace("$","").replace("%","").strip()
```

```python
h13.replace("\n","").replace("\t","").replace(":","").strip(" .,$!%")
```
