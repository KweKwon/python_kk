# 리스트(list)형 변수 - 데이터 분석에서 가장 많이 사용하는 자료형 1
* 모든 데이터 타입(정수, 실수, 문자열, 리스트. 딕셔너리)의 자료를 넣을 수 있다.
* 리스트 자료형은 순서가 있는 자료형이다.
* 순서가 있는 자료형은 index(번호)로 자료를 꺼내올 수 있다.
* 리스트 자료형은[자료1,자료2,자료3]형식으로 만든다.
* 리스트 자료형은 자료의 추가/삭제가 가능하다

```python
L =[1,2,3,"하이미디어",'파이썬',3.14]
```

```python
print(L)
type(L)
```

* 슬라이싱: 리스트에서 범위를 지정해 자료를 여러 개 가져오는 법
* 변수명[시작번호:끝번호 바로 앞까지]
* 변수명[시작번호:끝번호-1:간격]


```python
L[:]
```

```python
L[0:6]
```

```python
L[0:6:2]
```

```python
L[5:0:-1]
```

```python
L[:4]
```

```python
L[2:]
```

```python
L[::-2]
```

```python
L[::2] #first : = 몇번쨰부터 몇번쨰 까지 secend : = 얼마만큼 띄워서 가져와라.
```

```python
L2 = [1, 2, [3,4,5], 6, 7, 8]
```

# 딕셔너리형(Dictionary) JSON(제이슨) 형이라고도함 - 데이터 분석에서 가장 많이 사용되는 자료형2

* key : value 형태의 자료형으로 나중에 API를 이용해 자료를 수집하면 대부분 JSON 형태로 받게된다.
* 변수명 = (key : value, key2 : value2}
