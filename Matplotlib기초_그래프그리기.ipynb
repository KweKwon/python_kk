```python
np.random.seed(19680801)
data = { 'a': np.arange(50),
         'c': np.random.randint(0, 50, 50),
         'd': np.random.randn(50)
        }
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots(figsize=(5, 2.7), layout ='constrained')
ax.scatter('a','b',c='c',s='d',data=data) # 산점도 그래프 c= 찍히는 점의 색, s= 점의 크기
ax.set_xlabel('entry a')
ax.set_ylabel('entry b', rotation=0)
plt.show()
```

# Matplotlib의 2가지 코딩 스타일

## 1) 00-스타일 : matplotlib 공식 추천 스타일
* 복잡한 그래프를 그릴 때 사용
* 사용법이 pyplot 스타일 보다 복잡함

```python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots(figsize=(5,2.7), layout = 'constrained')
ax.plot(x,x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Sample Plot')
ax.legend()
plt.show()
```

## 2) pyplot 스타일
* 00-style보다 사용법이 간단하다
* 복잡한 그래프를 그릴때는 적합하지 않음

```python
x = np.linspace(0, 2, 100)

plt.figure(figsize=(5,2.7), layout = 'constrained')
plt.plot(x,x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Sample Plot')
plt.legend()
plt.show()
```

# 그래프 꾸미기

# 1) 선 모양 바꾸기 set_linestyle(모양)
* linestyle or ls: {'-', '--', '-.', ':', '', (offset, on-off-seq), ...}

```python
# 4개의 랜덤 데이터셋 만들기
data1, data2, data3, data4 = np.random.randn(4,100)
```

```python
fig, ax = plt.subplots(figsize=(5,2.7))
x = np.arange(len(data1))
ax.plot(x, np.cumsum(data1), color='blue', linewidth=3, linestyle='--')
l, = ax.plot(x, np.cumsum(data2), color='orange', linewidth=2)
l.set_linestyle(':')
plt.show()
```

## 2) 색 바꾸기 (facecolor, edgecolor)


```python
fig, ax = plt.subplots(figsize=(5,2.7))
ax.scatter(data1,data2,s=50, facecolor='C0', edgecolor='k')
plt.show()
```

## 3) 선 넓이, 선형태, 마커 모양과 크기

```python
fig, ax = plt.subplots(figsize=(5,2.7))
ax.plot(data1, 'o', label='data1')
ax.plot(data2, 'd', label='data2')
ax.plot(data3, 'v', label='data3')
ax.plot(data4, 's', label='data4')
ax.legend()
plt.show()
```

## 4) X축, Y축 이름, 눈금, 그래프 이름

```python
mu, sigma = 115,15
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
n, bins, patches = ax.hist(x, 50, density=True, facecolor='C0', alpha=0.75)

ax.set_xlabel('Legth [cm]')
ax.set_ylabel('parobablity')
ax.set_title('Aardvark lengths\n (not really)')
ax.text(75, .025, '$\mu=115,\ \sigma=15$')
ax.axis([55, 175, 0 , 0.03])
```

## 5) 주석, 부가설명 ax.annotate

```python
fig, ax = plt.subplots(figsize=(5,2.7))

t = np.arange(0.0,5.0,0.01)
s = np.cos(2 * np.pi * t)
line, = ax.plot(t,s, lw=2)

ax.annotate('local max', xy=(2,1), xytext=(3,1.5),arrowprops=dict(facecolor='black', shrink=0.05))
ax.set_ylim(-2,2)
plt.show()
```

## 7) 눈금자 scales

## 8) 눈금자 단위 구간 설정하기

# 9) 날짜와 문자 카테고리 표시하기

```python
fig, ax = plt.subplots(figsize = (5,2.7), layout='constrained')
dates = np.arange(np.datetime64('2021-11-15'), np.datetime64('2021-12-25'), np.timedelta64(1,'h'))
data = np.cumsum(np.random.randn(len(dates)))
ax.plot(dates, data)
cdf = mpl.dates.ConciseDateFormatter(ax.xaxis.get_major_locator())
ax.xaxis.set_major_formatter(cdf)
plt.show()
```
