다중 조건문 사용하기 elif

```
if 조건문:
    코드 (조건이 참일때 실행)
elif 조건문2
    코드 (조건2가 참일때 실행)
elif 조건문3
    코드 (조건3이 참일때 실행)
else:
    코드 (조건, 조건2, 조건3이 모두 거짓 일때)
```



학점 판별 프로그램<br>
81-100 : A<br>
61-60 : B<br>
41-60 : C<br>
21-40 : D<br>
0-20 : F


````python
score = int(input())
if 0 > score or score > 100:
  print ('범위를 벗어났습니다. 다시 입력해 주세요.')
else:
    if 80 < score <= 100:
      print('학점 A')
    elif 60 < score <= 80:
      print('학점 B')
    elif 40 < score <= 60:
      print('학점 C')
    elif 20 < score <= 40:
      print('학점 D')
    elif 0 <= score <=20:
      print('학점 F')
````

표준 입력으로 나이(만 나이)가 입력됩니다.(입력 겂은 7 이상입력됨)<br>
교통카드 시스템에서 시내버스 요금은 다음과 같으며 각 나이에 맞게 요금을 차감한 뒤 잔액이 출력되게 만드시오(if,elif사용)
현재 교통카드에는 9000원이 들어있습니다.<br>
어린이 (초등학생, 만 7세이상 12세 이하):650원<br>
청소년(중.고등학생, 만13세 이상 18세 이하):1050원<br>
어른(일반, 만 19세 이상):1250

````python
# 입력받는 곳(나이)
# 교통카드에 9000원 입력
# 나이가 7살 이상인지 판별
# 나이가 7살 이상이면 어린이, 청소년, 어른 별로 교통카드에서 요금 차감
# 잔액 출력
age = int(input('나이를 입력하세요:'))
card = 9000
if 7< age <= 100:
  if 7 <= age <= 12:
    print("잔액:",card-650,'원')
  elif 13 <= age <= 18:
    print("잔액:",card-1050,'원')
  elif 18 < age:
    print("잔액:",card-1250,'원')
else:
  print('7보다 크거나 100보다 작은 수를 입력하세요.')
````
